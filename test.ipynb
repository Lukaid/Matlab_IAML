```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

iris = pd.read_csv('iris.csv')
```

```python
iris['variety'][90]
```

```python
X_1 = iris['sepal.length'][0:100]
X_2 = iris['sepal.width'][0:100]
X = pd.concat([X_1, X_2], axis=1)
X
```

```python
Y = iris['variety'][0:100]
for i in range(0, 100):
    if Y[i] == 'Setosa':
        Y[i] = 1
    else:
        Y[i] = -1

Y
```

```python
plt.scatter(X['sepal.length'][0:50], X['sepal.width'][0:50], label = 'Setosa')
plt.scatter(X['sepal.length'][50:100], X['sepal.width'][50:100], label = 'Versicolor')
plt.xlabel('sepal.length')
plt.ylabel('sepal.width')
plt.legend()
plt.show()
```

```python
def step_f(z):
    if z >= 0:
        return 1
    else:
        return -1
```

```python
n_iter = 10
eta = 0.1
w = [0, 0, 0]

print(w)
```

```python
Yhat = []

for j in range(n_iter):
    for i in range(len(X)):
        Yhat.append(step_f(np.dot(w, [1] + list(X.loc[i]))))

        for k in range(0, 3):
            w[k] = w[k] + eta*(Y[i] - Yhat[i])*([1] + list(X.loc[i]))[k]
```

```python
x_1 = list(np.arange(4, 7, 0.1))
x_2 = []

for h in range(len(x_1)):
    x_2.append( -w[0]/w[2] -(w[1]/w[2])*x_1[h])


plt.plot(x_1, x_2)
plt.show()



#for h in range(len())
```
